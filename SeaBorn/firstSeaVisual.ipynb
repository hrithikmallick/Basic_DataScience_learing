```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#load dataset that are present in seaBorn 
fmri=sns.load_dataset("fmri")
fmri.head(10)
```

```python
#use lineplot sns methods
sns.lineplot(x="timepoint",y="signal",data=fmri)
plt.show()
```

```python
#Grouping data with help of HUE
sns.lineplot(x="timepoint",y="signal",data=fmri,hue="event")
plt.show()
```

```python
#Adding style on linePlot
sns.lineplot(x="timepoint",y="signal",data=fmri,hue="event",style="event")
plt.show()
```

```python
#adding markers
sns.lineplot(x="timepoint",y="signal",data=fmri,hue="event",style="event",markers=True)
plt.show()
```

```python
#Using lineplot with own dataset
heartData=pd.read_csv("heart.csv")
heartData.head(10)
```

```python
sns.lineplot(x="trtbps",y="chol",data=heartData,hue="exng",style="exng",markers=True)
plt.show()
```

```python
pokemonData=pd.read_csv("Pokemon.csv")
pokemonData.head()
```

```python
sns.barplot(x="is_legendary",y="weight_kg",data=pokemonData)
plt.show()
```

```python
#Another example of  hue in bar plot
sns.barplot(x="is_legendary",y="speed",hue="type1",data=pokemonData)
plt.show()
```

```python
#using differnt palette in bar plot
#Blues_d , rocket , vlag
sns.barplot(x="is_legendary",y="weight_kg",data=pokemonData,
palette="vlag")
plt.show()
```

```python
#using fixed color for both bars
sns.barplot(x="is_legendary",y="weight_kg",data=pokemonData,color="orange")
plt.show()
```

```python
#seaborn HIStogram/Distplot
#Its is used best for numercial columns

#add dimond dataset
diamondData=pd.read_csv("diamonds.csv")
diamondData.head(6)
```

```python

# IT will be remove in futere so new versions are kdeplot and rugplot
sns.distplot(diamondData["price"],rug=True)
plt.show()
```

```python

sns.distplot(diamondData["price"], hist=False)
plt.show()
```

```python
#rugplot new version
sns.rugplot(diamondData["price"])
plt.show()
```

```python
#hist false new version
sns.kdeplot(diamondData["price"])
plt.show()
```

```python
#Extra Atributes in distplot
sns.distplot(diamondData["price"],color="purple")
plt.show()
```

```python
#removing kde for frequency curve
sns.distplot(diamondData["price"],color="green",bins=10,kde=False)
plt.show()
```

```python
#change the axix of distplot from x_axis to y_axis
sns.distplot(diamondData["price"],color="orange",vertical=True)
plt.show()
```

```python
#Joint plot it is a mixture of scatter and distplor
#SeaBorn JointPlot
sns.jointplot(x="trtbps",y="chol",data=heartData)
plt.show()
```

```python
#adding other attributes in joint plot
sns.jointplot(x="trtbps",y="chol",data=heartData,color="brown")
plt.show()
```

```python
#kind attribute for regration line
sns.jointplot(x="trtbps",y="chol",data=heartData,kind="reg")
plt.show()
```

```python
#seaborn boxplot
#It is used for one character variable and one numbercial column (x,y)
diamondData.head()
```

```python
sns.boxplot(x="cut",y="price",data=diamondData)
plt.show()
```

```python
#palate in boxplot
sns.boxplot(x="cut",y="price",data=diamondData,palette="Set2")
plt.show()
```

```python
# adding color in boxplot
sns.boxplot(x="cut",y="price",data=diamondData,color="red")
plt.show()
```

```python
#Using hue parameter for box plotes
sns.boxplot(x="cut",y="price",data=diamondData,hue="clarity")
plt.show()
```
