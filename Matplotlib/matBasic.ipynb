```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# line plot
x=np.arange(1,11)
x
```

```python
y=x*5
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
#ading lable title in plot
plt.plot(x,y)
plt.title("LIne plot")
plt.xlabel("This is X-lable")
plt.ylabel("This is Y-lable")
plt.show()
```

```python
#plot attribute

plt.plot(x,y,color='r',linestyle=':',linewidth=5)
plt.show()
```

```python
#two plot
x=np.arange(1,11)
y1=2*x
y2=3*x
```

```python
plt.plot(x,y1,color='r',linestyle=":")
plt.plot(x,y2,color='y',linewidth=4)
plt.grid(True) #adding grid in line plot
plt.title("TWO line in a plot")
plt.xlabel("this is x axis")
plt.ylabel("this is y axis")
plt.show()
```

```python
#adding sub plots first is row second is column then no
plt.subplot(2,1,1)
plt.plot(x,y1,color='b',linestyle=":")
plt.subplot(2,1,2)
plt.plot(x,y2,color='y',linewidth=4)
plt.title("Two lines in a plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.grid(True)
plt.show()

```

```python
# Bar plot for categrorical distribution
stuedent={'Bob':187,'Maria':149,'anne':198,'hrk':135}
#extracting keys from dictanary
names=list(stuedent.keys())
names

```

```python
#exracting values from a dict
values=list(stuedent.values())
values
```

```python
#creating bar plot
plt.bar(names,values)
plt.show()
```

```python
#adding x and y lable and title
plt.bar(names,values,color="y")
plt.title("Bar Plot")
plt.xlabel("Names")
plt.ylabel("Values")
# plt.grid(True)
plt.show()
```

```python
#Horizontal Bar plot
plt.barh(names,values,color="g")
plt.title("Bar Plot")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

```python
#scatter plot
x=[4,7,8,6,3,4,8,1,9,2]
y=[5,7,8,9,4,6,1,2,3,4]
plt.scatter(x,y)
plt.show()
```

```python
#change marker
plt.scatter(x,y,marker="*",c="r",s=115)
plt.show()
```

```python
#Two marker in the same plot
x=[4,7,8,6,3,4,8,1,9,2]
y=[5,7,8,9,4,6,1,2,3,4]
z=[9,7,6,4,3,1,5,6,4,8]
plt.scatter(x,y,c="r",s=50)
plt.scatter(x,z,marker="*",c="g",s=50)
plt.show()
```

```python
#subplot
plt.subplot(1,2,1)
plt.scatter(x,y,c="r",s=50)
plt.subplot(1,2,2)
plt.scatter(x,z,marker="*",c="g",s=50)
plt.show()


```

```python
#Box-plot
one=[1,2,3,4,5,6,7,8,9]
two=[1,2,4,6,8,9,4,1,5]
three=[2,4,5,6,7,8,9,7,3]
data=list([one,two,three])
data
```

```python
#pie_chart
fruits=["Mango","Apple","Orange","Guava"]
quantity=[50,40,60,100]
plt.pie(quantity,labels=fruits)
plt.show()
```

```python
#adding  attributes in pie chart
plt.pie(quantity,labels=fruits,autopct="%0.1f",colors=["yellow","red","orange","green"])
plt.show()
```

```python
#DoughNut_Chart
fruits=["Mango","Apple","Orange","Guava"]
quantity=[4,12,18,20]
plt.pie(quantity,labels=fruits,radius=2)
plt.pie([100],radius=1,colors="white")
plt.show()
```
