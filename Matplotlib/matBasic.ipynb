```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# line plot
x=np.arange(1,11)
x
```

```python
y=x*5
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
#ading lable title in plot
plt.plot(x,y)
plt.title("LIne plot")
plt.xlabel("This is X-lable")
plt.ylabel("This is Y-lable")
plt.show()
```

```python
#plot attribute

plt.plot(x,y,color='r',linestyle=':',linewidth=5)
plt.show()
```

```python
#two plot
x=np.arange(1,11)
y1=2*x
y2=3*x
```

```python
plt.plot(x,y1,color='r',linestyle=":")
plt.plot(x,y2,color='y',linewidth=4)
plt.grid(True) #adding grid in line plot
plt.title("TWO line in a plot")
plt.xlabel("this is x axis")
plt.ylabel("this is y axis")
plt.show()
```

```python
#adding sub plots first is row second is column then no
plt.subplot(2,1,1)
plt.plot(x,y1,color='b',linestyle=":")
plt.subplot(2,1,2)
plt.plot(x,y2,color='y',linewidth=4)
plt.title("Two lines in a plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.grid(True)
plt.show()

```

```python
# Bar plot for categrorical distribution
stuedent={'Bob':187,'Maria':149,'anne':198,'hrk':135}
#extracting keys from dictanary
names=list(stuedent.keys())
names

```

```python
#exracting values from a dict
values=list(stuedent.values())
values
```

```python
#creating bar plot
plt.bar(names,values)
plt.show()
```

```python
#adding x and y lable and title
plt.bar(names,values,color="y")
plt.title("Bar Plot")
plt.xlabel("Names")
plt.ylabel("Values")
# plt.grid(True)
plt.show()
```
